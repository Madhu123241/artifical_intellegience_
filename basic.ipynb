```python
# Basic Data Types in Python

# Integer
age = 25
print(f"Integer: {age}, Type: {type(age)}")

# Float
height = 5.9
print(f"Float: {height}, Type: {type(height)}")

# String
name = "Alice"
print(f"String: {name}, Type: {type(name)}")

# Boolean
is_student = True
print(f"Boolean: {is_student}, Type: {type(is_student)}")

# List
fruits = ["apple", "banana", "cherry"]
print(f"List: {fruits}, Type: {type(fruits)}")

# Dictionary
person = {"name": "Bob", "age": 30}
print(f"Dictionary: {person}, Type: {type(person)}")

# Tuple
coordinates = (10, 20)
print(f"Tuple: {coordinates}, Type: {type(coordinates)}")
```

```python
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

if is_student:
    print(f"{name} is a student.")
else:
    print(f"{name} is not a student.")

if len(fruits) > 2:
    print("You have more than two fruits.")
elif len(fruits) == 2:
    print("You have exactly two fruits.")
else:
    print("You have one or no fruits.")

if person.get("age", 0) > age:
    print(f"{person['name']} is older than {name}.")
else:
    print(f"{person['name']} is not older than {name}.")
```
